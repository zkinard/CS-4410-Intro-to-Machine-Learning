```python
#a.: and b.:Download diamonds.csv and import dataset into Pandas DataFrame.
import pandas as pd
diamonds = pd.read_csv('https://vincentarelbundock.github.io/' +
                        'Rdatasets/csv/ggplot2/diamonds.csv', index_col=0)
#c: Display the first seven rows of the DataFrame
diamonds.head(7)
```

```python
#d: Display the last seven rows of the DataFrame
diamonds.tail(7)
```

```python
#Use Series the describe method (which looks only at the numerical columns)
#to calculate the descriptive statistics for the numerical columns
#—carat, depth, table, price, x, y, and z.
diamonds.describe()
```

```python
#Use Series method Describe to calculate the descriptive statistics
#for the categorical data (text) columns—cut, color, and clarity.
(diamonds.cut).describe()
```

```python
(diamonds.color).describe()
```

```python
(diamonds.clarity).describe()
```
